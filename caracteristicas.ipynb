```python
from facenet_pytorch import MTCNN
from facenet_pytorch import InceptionResnetV1
from PIL import Image
import pickle
import os

data_dir = '/home/fer/Descargas/rostros_'
save_dir = '/home/fer/Descargas/Repechaje'

model = InceptionResnetV1(pretrained='vggface2').eval()
mtcnn = MTCNN(min_face_size = 50, keep_all = False)
```

```python
inicio = 0
fin = 500

img_files = os.listdir(data_dir)[inicio:fin]
for img_file in img_files:
    img_path = os.path.join(data_dir, img_file)
    # check the file is an image or not
    if os.path.splitext(img_path)[1].lower() in ('.jpg', '.jpeg', '.png'):
    
            img = Image.open(img_path)
            img = img.convert("RGB")
            save_path = os.path.join(save_dir, os.path.splitext(img_file)[0]+'.png')
            # Obtener el nombre del archivo de la imagen actual
            
            face = mtcnn(img)
        
            if face == None:
                os.remove(img_path)
                img.save(save_path)
                
                
```

```python

inicio = 5500
fin = 5749
embeddings_list = []
Labels = []

img_files = os.listdir(data_dir)[inicio:fin]
for img_file in img_files:
    img_path = os.path.join(data_dir, img_file)
    # check the file is an image or not
    if os.path.splitext(img_path)[1].lower() in ('.jpg', '.jpeg', '.png'):
    
            img = Image.open(img_path)
            img = img.convert("RGB")
            save_path = os.path.join(save_dir, os.path.splitext(img_file)[0]+'.png')
            # Obtener el nombre del archivo de la imagen actual
            label = os.path.splitext(os.path.basename(img_path))[0]
            
            face = mtcnn(img)
        
            embeddings_list.append(model(face.unsqueeze(0)))
            Labels.append(label)
caracteristicas = dict(zip(Labels, embeddings_list))
len(caracteristicas)               
```

```python
list(caracteristicas.keys())
```

```python
filename = "./caracteristica_5500_5749" + ".pkl"
# Serializar la variable caracteristicas y guardarla en el archivo con el nombre generado
with open(filename, "wb") as f:
    pickle.dump(caracteristicas, f)
```

```python
#Unir varios diccionarios

folder_path = "./"
# Crear un diccionario vacío
result = {}

# Obtener la lista de nombres de archivo en la carpeta
file_names = os.listdir(folder_path)

# Iterar sobre la lista de nombres de archivo
for file_name in file_names:
    # Verificar que el archivo tenga la extensión .pkl
    if file_name.endswith(".pkl"):
        # Crear la ruta completa al archivo
        file_path = os.path.join(folder_path, file_name)
        # Abrir el archivo y cargar el diccionario
        with open(file_path, "rb") as f:
            data = pickle.load(f)
        # Actualizar el diccionario result con el diccionario cargado
        result.update(data)
# Crear el nombre del archivo 
filename = "./caracteristica.pkl"
# Serializar la variable caracteristicas y guardarla en el archivo con el nombre generado
with open(filename, "wb") as f:
    pickle.dump(result, f)
```
